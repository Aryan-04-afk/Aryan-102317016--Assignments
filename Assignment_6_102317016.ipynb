**Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.**

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)

y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$y = M \cdot x^2$', color='b', linestyle='--')
plt.plot(x, y2, label=r'$y = M \cdot \sin(x)$', color='r', linestyle='-')
plt.legend()
plt.grid(True)
plt.title("Plot of y = M * x^2 and y = M * sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()
```

**Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Score': [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
ax = sns.barplot(x='Subject', y='Score', data=df, palette='viridis')

for index, row in df.iterrows():
    ax.text(index, row.Score + 1, str(row.Score), ha='center', fontsize=12)

plt.title("Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

**Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.**

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102317016
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].plot(np.cumsum(data), color='b', linestyle='-', marker='o')
axes[0].set_title("Cumulative Sum Plot")
axes[0].set_xlabel("Index")
axes[0].set_ylabel("Cumulative Sum")
axes[0].grid(True)

noise = np.random.randn(50) * 0.5
axes[1].scatter(range(50), data + noise, color='r', marker='x')
axes[1].set_title("Scatter Plot with Noise")
axes[1].set_xlabel("Index")
axes[1].set_ylabel("Value + Noise")
axes[1].grid(True)

plt.tight_layout()

plt.show()
```


**Q4.Apply ‘seaborn’ library to do the following.
 Read Total profit of all months and show it using a line plot.
 Read all product sales data and show it using a multiline plot.
 Plot bar chart for all the features/attributes.**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files

uploaded = files.upload()

file = list(uploaded.keys())[0]
df = pd.read_csv(file)

sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.lineplot(x=df['month_number'], y=df['total_profit'], marker='o', color='b')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
for column in df.columns[1:-2]:
    sns.lineplot(x=df['month_number'], y=df[column], label=column)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
df.set_index('month_number').plot(kind='bar', stacked=False, colormap='viridis', width=0.8)
plt.title("Sales Data Bar Chart")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend(title="Products", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='y')
plt.show()
```

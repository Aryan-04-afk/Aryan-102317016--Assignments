```python
import numpy as np
gfg=np.matrix('[4,1,9;12,3,1;4,5,6]')


total=gfg.sum()
print("The sum of all elements in the matrix is: ",total)

row_wise_total=gfg.sum(axis=1)
print("The row wise sum of matrix is : ",row_wise_total)

coloum_wise_total=gfg.sum(axis=0)
print("The coloumn wise sum of matrix is: ",coloum_wise_total)
```

```python
array=np.array([10,52,62,16,16,54,453])
sorted_array=np.sort(array)

print("The sorted array is: ",sorted_array)

indices=np.argsort(array)
print("The indices of the sorted array is: ",indices)


arr=np.array(array)

smallest_4=np.partition(arr,4)[:4]
greatest_5=np.partition(arr,-5)[-5:]

print("The smallest 4 elements in the array is :", smallest_4)
print("The greatest 5 elements in the array is :", greatest_5)

arr2=np.array([1.0,1.2,2.2,2.0,3.0,2.0])
integer=arr2[arr2==arr2.astype(int)]
floats=arr2[arr2!=arr2.astype(int)]

print("The integer elements in the array are: ",integer)
print("The float elements in the array are: ",floats)



```

```python
sales=np.array([670,720,770,820,870])
personalized_tax_rate=(((sales%5) + 5) /100)




new_sales = sales * (1 + personalized_tax_rate)

print("The new sales array is: ", new_sales)

x=new_sales[0]+100


new_sales=np.where(new_sales<x,new_sales*0.95,new_sales*0.9)


print("The new sales array after discount is :",new_sales)


sales_matrix=np.tile(sales,(3,1))
print("The sales matrix is : ",sales_matrix)


for i in range(1, sales_matrix.shape[0]):
    sales_matrix[i] = sales_matrix[i-1] * 1.02  # Increase by 2%

print("The new sales matrix after 2% week_wise increase is",sales_matrix)
```

```python
import matplotlib.pyplot as plt
array=np.linspace(-10,10,100)

y_square=array*array

plt.plot(array,y_square)
plt.title("Plot of x square")
plt.xlabel("x")
plt.ylabel("x^2")
plt.grid(True)

plt.show()

y_sin=np.sin(array)
plt.plot(array,y_sin)
plt.title("Plot of sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)

plt.show()


y_exp=np.exp(array)

plt.plot(array,y_exp)
plt.title("Plot of e^x")
plt.xlabel("x")
plt.ylabel("e^x")
plt.grid(True)

plt.show()


y_log=np.log(np.abs(array)+1)
plt.plot(array,y_log)
plt.title("Plot of log(x)")
plt.xlabel("x")
plt.ylabel("log(x)")
plt.grid(True)

plt.show()

```
